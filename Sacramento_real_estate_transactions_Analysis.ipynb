```python
# Using Pandas.
import pandas as pd
```

```python

# 1. Read the csv file into python
df = pd.read_csv('/content/drive/MyDrive/Adv Python for Data Analyst/Week 2/HW 1/Sacramentorealestatetransactions.csv')
display(df.head(5))
# 9. Write a short description of the data set
text1 = """This data contains records of 985 real estate transactions
in the Sacramento, California area. \n It provides a detailed summary
of property sales, such as location, property features, and transaction details."""
display(text1)
```

```python
# 2. Describe() the data
df.describe(include='all')
```

```python
# 3. Create two new datasets using crosstabs from the data set
cross_1 = pd.crosstab(df['city'],df['type'])
cross_2 = pd.crosstab(df['price'],df['city'])
```

```python
# Print the crosstabs
display(cross_1.sort_values(by='Residential', ascending=False))
# SACRAMENTO has the most Condos,	Multi-Families, and	Residentials
display(cross_2)
```

```python
# 4. Describe() the results of the crosstabs
display(cross_1.describe(include='all'))
text2 = """ The "Residential" is the most common property type across
cities, accounting for the majority of all transactions. \n Sacramento
show a diverse mix of property types, including 27 Condos and 10
Multi-Family, and 402 Residentials. \n On the other hand, areas, like
Elk Grove or Lincoln, are almost exclusively Residential, showing very
little variation in property type."""
print(text2)

display(cross_2.describe(include='all'))
text3 = """Most transactions range between $100,000 and $300,000, which
represents the most active price bracket for buyers in this dataset.
\n A few transactions consist of properties selling for over $500,000,
indicating that luxury properties are concentrated in specific neighborhoods like Granite Bay."""
print (text3)
```

```python
# 6. Slice a series of rows from the dataset based on a value from one of the columns
Sliced_df = df[df['city'] == 'SACRAMENTO']
display(Sliced_df)
```

```python
# 7. Describe the new dataset
display(Sliced_df.describe())
```

```python
import matplotlib.pyplot as plt
# 8. Slice out a row or two of data from the crosstab to graph and then graph the results - don't forget to lable your graphs
sliced_cross_1 = cross_1.loc['SACRAMENTO']
sliced_cross_1.plot(kind='pie', figsize=(12, 6), autopct='%1.1f%%', labels=sliced_cross_1.index)
plt.title('Sacramento`s Neighborhood')
plt.savefig('sacramento_neighborhood_pie_chart.png')

text4 ="""This pie chart breakdowns property types within the city of
Sacramento. The graph highlights key insights: \n The chart shows that
Residential properties are majority of the local market (approximately
91.6%), representing the primary neighborhood profile for this area.
\n It captures the presence of Condos (approx. 6.2%) and Multi-Family
units (approx. 2.3%)."""

print(text4)
```
